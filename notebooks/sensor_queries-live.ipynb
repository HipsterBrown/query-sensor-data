```python
import os
from dotenv import load_dotenv
from viam.rpc.dial import DialOptions
from viam.robot.client import RobotClient
from viam.app.viam_client import ViamClient
from viam.utils import create_filter

load_dotenv()

async def get_machine_client() -> ViamClient:
    opts = RobotClient.Options.with_api_key(
        api_key=os.environ.get("VIAM_API_KEY"),
        api_key_id=os.environ.get("VIAM_API_KEY_ID"),
    )
    return await RobotClient.at_address(os.environ.get('ROBOT_ADDRESS'), opts)
```

```python
from datetime import datetime
import time
from matplotlib.animation import FuncAnimation
from bokeh.io import output_notebook, show, push_notebook
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from viam.components.movement_sensor import MovementSensor

output_notebook()

machine_client = await get_machine_client()
washer_movement = MovementSensor.from_robot(machine_client, os.environ.get("MOVEMENT_SENSOR_NAME"))

sensor_data = await washer_movement.get_readings()
display(sensor_data)

data_source = ColumnDataSource(data = {"Accel X": [sensor_data['linear_acceleration'].x], "DateTime": [datetime.now()]})

fig = figure(x_axis_type="datetime",
             width=900, height=450,
             tooltips=[("Accel X", "@AccelX"),],
             title="Accel X Live (Every Second)"
            )

fig.line(x="DateTime", y="Accel X", line_color="tomato", line_width=3.0, source=data_source)

fig.xaxis.axis_label = "Date Time"
fig.yaxis.axis_label = "Accel X"

handle_line_chart = show(fig, notebook_handle=True)

while True:
    data = await washer_movement.get_readings()
    new_row = {"Accel X": [data['linear_acceleration'].x,], "DateTime": [datetime.now(),]}
    data_source.stream(new_row)
    push_notebook(handle=handle_line_chart)
    time.sleep(1)
```
